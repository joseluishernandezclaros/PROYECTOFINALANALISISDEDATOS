# VERIFICAR TIPO DE DATO DE LAS COLUMNAS Y CONVERTIRLAS AL TIPO DE DATO CORRECTO.


```python
df = pd.read_csv('housing_fincaraiz_graf.csv', sep=";")
df.info()
```

```python

# Supongamos que tu DataFrame se llama 'df'

# Convierte las columnas a numérico, maneja valores no esperados
df['habitaciones'] = pd.to_numeric(df['habitaciones'], errors='coerce')
df['baños'] = pd.to_numeric(df['baños'], errors='coerce')
df['parqueaderos'] = pd.to_numeric(df['parqueaderos'], errors='coerce')
df['precio'] = pd.to_numeric(df['precio'], errors='coerce')

# Convierte "No definida" a NaN
df.replace('No definida', pd.NA, inplace=True)

# Rellena los valores NaN con 0
df.fillna(0, inplace=True)

# Convierte 'area_construida' y 'area_privada' a numérico, maneja valores no esperados
df['area_construida'] = pd.to_numeric(df['area_construida'].astype(
    str).str.replace(' m²', ''), errors='coerce')
df['area_privada'] = pd.to_numeric(df['area_privada'].astype(
    str).str.replace(' m²', ''), errors='coerce')

# Convierte 'area_construida' y 'area_privada' a Int64
df['area_construida'] = df['area_construida'].round().astype('Int64')
df['area_privada'] = df['area_privada'].round().astype('Int64')

# Luego, convierte las demás columnas a int64
df['habitaciones'] = df['habitaciones'].astype('int64')
df['baños'] = df['baños'].astype('int64')
df['parqueaderos'] = df['parqueaderos'].astype('int64')
df['precio'] = df['precio'].astype('int64')

print("TIPOS DE DATOS MODIFICADOS")
print(df.info())
```

# PRECIO PROMEDIO POR CANTIDAD DE HABITACIONES


```python
import pandas as pd

df = pd.read_csv('housing_fincaraiz_graf.csv', sep=";")
a = df['precio']
b = df['habitaciones']

# Supongamos que df2 es tu DataFrame y tiene las columnas 'precio' y 'habitaciones'
# a = df2['precio']
# b = df2['habitaciones']

# Crear un nuevo DataFrame con las columnas 'precio' y 'habitaciones'
df = pd.DataFrame({'precio': a, 'habitaciones': b})

# Convertir las columnas a tipos numéricos si no lo están ya
df['precio'] = pd.to_numeric(df['precio'], errors='coerce')
df['habitaciones'] = pd.to_numeric(df['habitaciones'], errors='coerce')

# Eliminar filas con valores nulos
df = df.dropna()

# Agrupar por el valor de 'habitaciones'
grouped = df.groupby('habitaciones')

# Inicializar listas para almacenar los resultados
habitaciones_values = []
promedio_precios = []

# Iterar sobre los grupos y calcular el promedio del precio para cada grupo de habitaciones
for habitaciones, group in grouped:
    habitaciones_values.append(habitaciones)
    promedio_precio = group['precio'].sum() / len(group)
    promedio_precios.append(promedio_precio)

# Crear un nuevo DataFrame con los resultados
resultados = pd.DataFrame(
    {'habitaciones': habitaciones_values, 'promedio_precio': promedio_precios})

# Ordenar los resultados por la cantidad de habitaciones de manera ascendente
resultados = resultados.sort_values(by='habitaciones')

# Mostrar el DataFrame con los resultados
print(resultados)

plt.figure(figsize=(10, 6))
plt.plot(resultados.index, resultados.values, marker='o', linestyle='-')
plt.xlabel('Cantidad de Habitaciones')
plt.ylabel('Promedio del Precio')
plt.title('Promedio del Precio por Cantidad de Habitaciones')
plt.grid(True)
plt.show()
```
