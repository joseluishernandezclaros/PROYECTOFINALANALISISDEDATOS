```python
# Reemplazar 'No definida' con ceros
df2['habitaciones'] = df2['habitaciones'].replace('No definida', 0)

# Convertir la columna 'habitaciones' a tipo numérico (entero)
df2['habitaciones'] = df2['habitaciones'].astype(int)

# Forzar la conversión de la columna 'precio' a cadena de caracteres (str)
df2['precio'] = df2['precio'].astype(str)

# Formatear el precio (eliminar comas y ceros a la derecha)
df2['precio'] = df2['precio'].str.replace(',', '', regex=False).str.rstrip(
    '0').str.replace('.', '', regex=False).str.replace(',', '.', regex=False).astype(float)

# Reemplazar valores no finitos (NaN o inf) en la columna 'habitaciones' por ceros
df2['habitaciones'] = df2['habitaciones'].fillna(0)

# Convertir la columna 'habitaciones' a tipo numérico (entero)
df2['habitaciones'] = df2['habitaciones'].astype(int)

# Agrupar por la cantidad de habitaciones y calcular el promedio del precio
resultados = df2.groupby('habitaciones')['precio'].mean()

# Ordenar los resultados por la cantidad de habitaciones de manera ascendente
resultados = resultados.sort_index(ascending=True)

# Configurar el formato de impresión para números
# Cambia el formato a dos decimales
pd.options.display.float_format = '{:.2f}'.format

# Mostrar los resultados
print(resultados)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(resultados.index, resultados.values, marker='o', linestyle='-')
plt.xlabel('Cantidad de Habitaciones')
plt.ylabel('Promedio del Precio')
plt.title('Promedio del Precio por Cantidad de Habitaciones')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(resultados.index, resultados.values)
plt.xlabel('Cantidad de Habitaciones')
plt.ylabel('Promedio del Precio')
plt.title('Promedio del Precio por Cantidad de Habitaciones')
plt.show()
```

```python
# Filtrar las filas que no sean "No definida" y que no sean 0 habitaciones
df2 = df2[(df2['habitaciones'] != "No definida") & (df2['habitaciones'] != 0)]

# Convertir la columna "habitaciones" a tipo entero
df2['habitaciones'] = df2['habitaciones'].astype(int)

# Agrupar por la columna "habitaciones"
grupos = df2.groupby('habitaciones')

# Contar cuántos elementos hay en cada grupo
conteo = grupos.size()

# Ordenar en orden ascendente por la columna "habitaciones"
conteo = conteo.sort_index()

# Mostrar el resultado
print(conteo)

# Graficar los resultados
conteo.plot(kind='bar')
plt.xlabel('Habitaciones')
plt.ylabel('Conteo')
plt.title('Conteo de Habitaciones')
plt.show()
```

```python


# Eliminar filas con 'No definida' en la columna 'baños'
df2 = df2[df2['baños'] != 'No definida']

# Convertir la columna "baños" a tipo entero
df2['baños'] = df2['baños'].astype(int)

# Agrupar por la columna "baños"
grupos = df2.groupby('baños')

# Contar cuántos elementos hay en cada grupo
conteo = grupos.size()

# Ordenar en orden ascendente por la columna "baños"
conteo = conteo.sort_index()

# Mostrar el resultado
print(conteo)

# Graficar los resultados
conteo.plot(kind='bar')
plt.xlabel('Baños')
plt.ylabel('Conteo')
plt.title('Conteo de Baños')
plt.show()
```
